```python
df = pd.read_csv('WorldCups.csv')
df.describe()
df.head()
```

```python

sns.barplot(x = df['Year'].astype(str), y = df['Attendance'], data = df)
plt.xticks(rotation=45)
```

```python
sns.countplot(data = df, x = df['Winner'])
first = df['Winner'].value_counts(dropna = False)

second = df['Runners-Up'].value_counts(ascending=False)
third = df['Third'].value_counts(ascending=False)
plt.xticks(rotation=45);

first
```

```python
winnings = first.add(second, fill_value=0).add(third, fill_value=0)
winnings.sort_values(ascending=False, inplace=True)
sns.barplot(winnings)
plt.xticks(rotation=90);
```
