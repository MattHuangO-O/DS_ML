```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
scaler = StandardScaler()
edu_scaled = pd.DataFrame(scaler.fit_transform(edu), columns = edu.columns)
edu_scaled
```

# PCA

```python
pca = PCA(n_components=edu.shape[1], random_state=1) #need to input number of components into PCA
edu_pca = pd.DataFrame(pca.fit_transform(edu_scaled))
exp = pca.explained_variance_ratio_
exp.cumsum()
```

```python
plt.figure(figsize=(10,10))
plt.plot(exp.cumsum(), marker = 'o', linestyle = "--")
plt.xlabel("Principal Components")
plt.ylabel("Explained Variance Ratio")
plt.show()
```

```python
i = 1
for x in exp.cumsum():
    if x > .70:
        print("Number of Principal Components for 90% Variation is: ", i)
        break
    else:
        i += 1
```

```python
pc_comps = ['PC1', 'PC2', 'PC3', 'PC4']
edu_pca_4 = pd.DataFrame(pca.components_[:4,:], index = pc_comps, columns = edu_scaled.columns)
edu_pca_4.T
pca.components_
```

```python
def color_high(val):
    if val <= -0.24: # you can decide any value as per your understanding
        return 'background: green'
    elif val >= 0.3:
        return 'background: blue'   
    
edu_pca_4.T.style.applymap(color_high)
```
